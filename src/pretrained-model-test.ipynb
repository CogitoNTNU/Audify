```python
%pip install --upgrade transformers
%pip install "numpy<2.0"
%pip show transformers
from transformers import pipeline


# Defining a text-to-speech pipeline using a pretrained model suno/bark-small
pipe = pipeline("text-to-speech", model="suno/bark-small")
```

```python
# Generate a text-to-speech of a given text
text = "Suzanne loves numpy and Klein. Sharuca loves pandas and matplotlib. Nathania and Muhammad love computers and programming."
output = pipe(text)
```

```python
from IPython.display import Audio

Audio(output["audio"], rate=output["sampling_rate"])
```
